```python
# Computing count of variables in all columns to check for bias 
def count(name):
  x=df[name].unique()
  y=list(range(len(x)))
  for j in range(len(x)):
    y[j]=0
  for i in range(df.shape[0]):
    for j in range(len(x)):
      if(df[name][i]==x[j]):
        y[j]+=1
  for i in range(len(x)):
    print(x[i],"==",y[i])
```

```python
one_hot_encoded_data = pd.get_dummies(df1, columns = ['segment', 'model', 'fuel_type',
       'max_torque', 'max_power', 'engine_type', 'is_adjustable_steering',
       'is_parking_camera', 'transmission_type', 'steering_type',
       'is_front_fog_lights', 'is_rear_window_defogger', 'is_brake_assist',
       'is_driver_seat_height_adjustable', 'is_day_night_rear_view_mirror'])
```

```python
one_hot_encoded_data
```

```python
df2 = one_hot_encoded_data
```

```python
#Input and Output Split
y = df2.iloc[:,5]
```

```python
df2.drop('is_claim', axis=1, inplace=True)
```

```python
x = df2
```

```python
df2.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators = 100,
                               criterion="gini",
                               min_samples_split=5,
                               min_samples_leaf=3,
                               random_state=42)
model.fit(x,y)
```

```python
## Evaluation of Test dat (Prediction)
y_pred = model.predict(x)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
# Train Data Accuracy
accuracy_score(y, model.predict(x))
```

```python
result1 = classification_report(y, y_pred)
print("Classification Report:",)
print (result1)
```
